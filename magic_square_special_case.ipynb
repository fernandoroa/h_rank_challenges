```python
# Different to the h rank challenge this code
# allows to find solutions with less cost but repeated digits

from itertools import permutations
from itertools import product

def formingMagicSquare(s):
  # split natural number in components
  # a sum of components produce the natural number
  # produces list of tuples
  def decompose(k, n):
    if k == 1:
        return [(n,)]
    comp = []
    for i in range(0, n + 1):
        for t in decompose(k - 1, n - i):
            comp.append((i,) + t)
    no_dups = [t for t in (set(tuple(sorted(i)) for i in comp))]
    return no_dups
  
  # transforms list of tuples into lists
  # removes zeros
  # remove sublists with invalid numbers >= 9
  # remove invalid sublist lengths > 9
  def clean_compose(comp):
      comp_no_dups = [list(ele) for ele in comp]
      lst = list(map(lambda l: [v for v in l if v != 0], comp_no_dups))
      filtered_lst = []
      for sublist in lst:
          if max(sublist) <= 8 and len(sublist) <= 9:
              filtered_lst.append(sublist)
      return filtered_lst
  
  # gets stats from compositions, see above
  def make_stats(composition):
    size = len(composition)
    list_indexes = list(range(0, size))
    mydict = dict.fromkeys(list_indexes)
    idx = 0
    for item in composition:
      mydict[idx] = {'len': len(item), 
                    'len_unique': len(set(item)),
                    'lengths': {i:composition[idx].count(i) for i in set(composition[idx])}
                    }
      idx += 1
    return mydict
  
  # appends zeros to compositions to get lists
  # of length 9
  def append_to_comb(mylist, count_of_zeros):
    mylist_copy = [x[:] for x in mylist]
    to_append = [0] * count_of_zeros
    for idx in range(len(mylist)):
      mylist_copy[idx].extend(to_append)
    return mylist_copy

  # produce permutations of the possible solutions
  # sum putative solutions to original array (original square)
  def randomize(number_list, count_of_zeros, flat_list):
    steps = append_to_comb(number_list, count_of_zeros)

    array_accum = []
    for step in steps:
      test_list = list(set(permutations(step)))    
      res = [list(ele) for ele in test_list]
      for array1 in res:
        sum1 = [sum(x) for x in zip(flat_list, array1)]
        if all(i >= 1 and i <= 9 for i in sum1):
          check_equal = sum_checker(sum1)
          if check_equal:
            print("Magic Array Solution Found!!")
            return [True, cost]

      array_accum.extend(res)    
    return [False, cost]
  
  
  base_one_list = list([[1], [-1]])
  
  # sum to check if rows columns, diagonals sum the same
  def sum_checker(flat_list):
    row1 = sum(flat_list[0:3])
    row2 = sum(flat_list[3:6])
    row3 = sum(flat_list[6:9])
    col1_idx = [0, 3, 6]
    col2_idx = [1, 4, 7]
    col3_idx = [2, 5, 8]
    col1 = sum(list(map(flat_list.__getitem__, col1_idx)))
    col2 = sum(list(map(flat_list.__getitem__, col2_idx)))
    col3 = sum(list(map(flat_list.__getitem__, col3_idx)))
    diag1_idx = [0, 4, 8]
    diag2_idx = [2, 4, 6]
    diag1 = sum(list(map(flat_list.__getitem__, diag1_idx)))
    diag2 = sum(list(map(flat_list.__getitem__, diag2_idx)))
    if row1 == row2 == row3 == col1 == col2 == col3 == diag1 == diag2:
      print(' '.join(["Magic Number", str(row1)]))
      return True
    else:
      return False
    
  # get positive/negative combinations for possible solutions
  def getCombinations(x, places):
    return [list(p) for p in product(x, repeat=places)]

  flat_list = [x for xs in s for x in xs]
  status = False
  cost = 0
  while status == False and cost <= 30:
      cost += 1
      print(' '.join(["cost:", str(cost)]))
      comp = decompose(cost, cost)
      clean_components = clean_compose(comp)
      print(clean_components)

      compon_stats = make_stats(clean_components)
      
      print(' '.join(["stats:", str(compon_stats)]))      
      print(' '.join(["len(clean_components)", str(len(clean_components))]))
      print(' '.join(["compon_stats:", str(len(compon_stats))]))

      for current_index in range(len(clean_components)):
        print(' '.join(["current_index:", str(current_index)]))

        places = compon_stats[current_index]["len"]

        if places == 1:
          L_base = clean_components[current_index][0]
          base_list = [[z * L_base for z in x] for x in base_one_list]
          status_and_cost = randomize(base_list, 9 - places, flat_list)
          status = status_and_cost[0]
        else:
          combinations_one_base = getCombinations([1,-1], places)

          current_array = clean_components[current_index]

          combined_array = [[y * current_array[yi] for yi, y in enumerate(x)] for xi, x in enumerate(combinations_one_base) ]

          status_and_cost = randomize(combined_array, 9 - len(current_array), flat_list)
          status = status_and_cost[0]
          
        if status:
          return status_and_cost[1]

        
```
